**Exercise 2:** Map the **sqft_living** and **price** attributes to x and y axes of a scatter plot, respectively. Use some alpha transparency.

```r
# YOUR CODE HERE
library(ggplot2)

# Scatter plot in R
ggplot(County_data, aes(x = sqft_living, y = price)) +
  geom_point(alpha = 0.5) +
  labs(x = "Square Footage (sqft_living)", y = "Price") +
  ggtitle("Scatter Plot of sqft_living vs. Price")
```

We can't really see anything here, let's try the same plot with log scales. 

**Exercise 3:** Plot the above axes (**both x and y**) in log10 scale. 

```r
# YOUR CODE HERE
# Scatter plot with log10 scale in R
ggplot(County_data, aes(x = log10(sqft_living), y = log10(price))) +
  geom_point(alpha = 0.5) +
  labs(x = "Log10 Square Footage (log10(sqft_living))", y = "Log10 Price") +
  ggtitle("Scatter Plot of Log10 sqft_living vs. Log10 Price")
```

This gives a better relation between price and square footage. 


Let's plot a histogram of price; it is the distribution of prices for all houses. 

**Exercise 4:** Plot a **histogram** of **price** with a **binwidth** of 30000.

```r
# YOUR CODE HERE
# Histogram in R
ggplot(County_data, aes(x = price)) +
  geom_histogram(binwidth = 30000, fill = "blue", color = "black", alpha = 0.7) +
  labs(x = "Price", y = "Frequency") +
  ggtitle("Histogram of Price with Binwidth = 30000")
```

We can also plot the probability density function of the price, it will look like a smoothed version of the histogram.

**Exercise 5:** Plot the density of **price** using **geom_density**. 

```r
# YOUR CODE HERE
# Density plot in R
ggplot(County_data, aes(x = price)) +
  geom_density(fill = "green", color = "black", alpha = 0.7) +
  labs(x = "Price", y = "Density") +
  ggtitle("Density Plot of Price")
```

We can plot **multiple densities** on the same plot to see how the price distribution differs with respect to some attributes. We should use alpha transparency in `geom_density` to see the different distributions.

**Exercise 6:** Plot multiple densities of **price** with respect to **number of bedrooms**. 

```r
# YOUR CODE HERE
# Multiple densities with respect to number of bedrooms in R
ggplot(County_data, aes(x = price, fill = factor(bedrooms))) +
  geom_density(alpha = 0.7) +
  labs(x = "Price", y = "Density") +
  ggtitle("Densities of Price with Respect to Number of Bedrooms") +
  facet_wrap(~bedrooms, scales = "free_y")

```

This works well if there aren't too many classes. Here, we can't see much, so let's use the **facet_wrap()** to create **small multiples** to compare densities for different number of bedrooms. 

**Exercise 7:** Plot **small multiples** of **price densities** with respect to **number of bedrooms**. 

```r
# YOUR CODE HERE
# Small multiples of price densities with respect to number of bedrooms in R
ggplot(County_data, aes(x = price, fill = factor(bedrooms))) +
  geom_density(alpha = 0.7) +
  labs(x = "Price", y = "Density") +
  ggtitle("Densities of Price with Respect to Number of Bedrooms") +
  facet_wrap(~bedrooms, scales = "free_y", ncol = 3)
```

Here we can see that up to four bedrooms, the price distribution is narrow; after that the variance of the price increases. 
Let's see the distribution for number of floors.

**Exercise 8:** Plot **small multiples** of **price densities** with respect to **number of floors**. 

```r
# YOUR CODE HERE
# Small multiples of price densities with respect to number of floors in R
ggplot(County_data, aes(x = price, fill = factor(floors))) +
  geom_density(alpha = 0.7) +
  labs(x = "Price", y = "Density") +
  ggtitle("Densities of Price with Respect to Number of Floors") +
  facet_wrap(~floors, scales = "free_y", ncol = 3)
```

Let's plot a scatter plot for price vs. square footage using small multiples for number of bedrooms. Map the number of floors to *color* visual variable using a **sequential** brewer palette.

**Exercise 9:** Do a scatter plot of **price** vs. **square footage** with **small multiples** of **bedrooms**. 

```r
# YOUR CODE HERE
# Scatter plot with small multiples for number of bedrooms and color by number of floors in R
ggplot(County_data, aes(x = sqft_living, y = price, color = factor(floors))) +
  geom_point(alpha = 0.7) +
  labs(x = "Square Footage (sqft_living)", y = "Price") +
  ggtitle("Scatter Plot of Price vs. Square Footage with Bedrooms (Color by Floors)") +
  facet_wrap(~bedrooms, scales = "free_y") +
  scale_color_brewer(palette = "Blues")
```

**Exercise 12:** Plot a scatter plot of **gender** vs. **eyecolor**.

```r
# YOUR CODE HERE

library(ggplot2)

# Scatter plot of gender vs. eyecolor
ggplot(ecg, aes(x = gender, y = eyecolor)) +
  geom_point() +
  labs(x = "Gender", y = "Eyecolor") +
  ggtitle("Scatter Plot of Gender vs. Eyecolor")
```

This didn't work well. When we have overplotting problem where attributes have same exact values for a large number of data rows, we should use geom_jitter() to randomly place points in a scatter plot.

**Exercise 13:** Plot a scatter plot of **gender** vs. **eyecolor** with **jitter** geometry. 

```r
# YOUR CODE HERE

library(ggplot2)

# Scatter plot of gender vs. eyecolor with jitter
ggplot(ecg, aes(x = gender, y = eyecolor)) +
  geom_jitter() +
  labs(x = "Gender", y = "Eyecolor") +
  ggtitle("Scatter Plot of Gender vs. Eyecolor with Jitter")
```

It's intuitive to use **color** visual variable for the **eyecolor** attribute.

**Exercise 14:** Plot a scatter plot of **gender** vs. **eyecolor** and use **color** for the **eyecolor** attribute. Adjust width and height of **jitter** suitably and add transparency. 

```r
# YOUR CODE HERE

library(ggplot2)

# Scatter plot of gender vs. eyecolor with color and jitter
ggplot(ecg, aes(x = gender, y = eyecolor, color = eyecolor)) +
  geom_jitter(width = 0.2, height = 0.2, alpha = 0.7) +
  labs(x = "Gender", y = "Eyecolor") +
  ggtitle("Scatter Plot of Gender vs. Eyecolor with Color and Jitter")
```

The colors should be intuitive, so we will manually name them with **scale_color_manual()** where the color values will be 
c("blue", "chocolate4", "green4", "#595c26", "black").

**Exercise 15:** Add manual colors to the above plot.

```r
# YOUR CODE HERE

library(ggplot2)

# Define manual colors
manual_colors <- c("blue", "chocolate4", "green4", "#595c26", "black")

# Scatter plot of gender vs. eyecolor with manual colors and jitter
ggplot(ecg, aes(x = gender, y = eyecolor, color = eyecolor)) +
  geom_jitter(width = 0.2, height = 0.2, alpha = 0.7) +
  scale_color_manual(values = manual_colors) +  # Add manual colors
  labs(x = "Gender", y = "Eyecolor") +
  ggtitle("Scatter Plot of Gender vs. Eyecolor with Manual Colors and Jitter")
```

**Exercise 16:** And finally add **shape** visual variable to encode the **exercise** attribute. **Does it work well? Why?** 


```r
# YOUR CODE HERE

library(ggplot2)

# Define manual colors and shapes
manual_colors <- c("blue", "chocolate4", "green4", "#595c26", "black")
manual_shapes <- c(16, 17, 18, 19, 20)  # Adjust shapes as needed

# Scatter plot of gender vs. eyecolor with color, manual colors, shape, and jitter
ggplot(ecg, aes(x = gender, y = eyecolor, color = eyecolor, shape = exercise)) +
  geom_jitter(width = 0.2, height = 0.2, alpha = 0.7) +
  scale_color_manual(values = manual_colors) +
  scale_shape_manual(values = manual_shapes) +  # Add shape encoding
  labs(x = "Gender", y = "Eyecolor") +
  ggtitle("Scatter Plot of Gender vs. Eyecolor with Color, Manual Colors, Shape, and Jitter")
```

**YOUR ANSWER HERE**


```r
# 'exercise' has too many categories, shape encoding might not be effective.
# Additionally, there is overlap or ambiguity in the shapes, it does not work well for distinguishing categories.
```

**Exercise 17:** Plot a scatter plot of **gender** vs. **height** with **small multiples** for **exercise**. 

**Use the techniques you applied in the above exercises. Use intuitive visual variables for the attributes.** 


```r
# YOUR CODE HERE

library(ggplot2)

# Define manual colors and shapes
manual_colors <- c("blue", "chocolate4", "green4", "#595c26", "black")
manual_shapes <- c(16, 17, 18, 19, 20)  # Adjust shapes as needed

# Scatter plot of gender vs. height with small multiples for exercise
ggplot(ecg, aes(x = gender, y = height, color = eyecolor, shape = exercise)) +
  geom_jitter(width = 0.2, height = 0.2, alpha = 0.7) +
  scale_color_manual(values = manual_colors) +
  scale_shape_manual(values = manual_shapes) +  # Add shape encoding
  facet_wrap(~exercise) +  # Small multiples for exercise
  labs(x = "Gender", y = "Height") +
  ggtitle("Scatter Plot of Gender vs. Height with Small Multiples for Exercise")
```
